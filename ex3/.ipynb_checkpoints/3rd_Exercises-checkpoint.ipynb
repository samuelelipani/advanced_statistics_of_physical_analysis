# Exercise 1

```r
dtriangular <- function(x) { tr <- ifelse(2 <= x & x < 4,
                                      2 * (x - 2) / ((6 - 2)*(4 - 2)),
                                      ifelse(4 <= x & x <= 6, 
                                            2 * (6 - x)/((6-2)*(6-4)),
                                            0))
    return (tr)}

ptriangular <- function(x) { tr <- ifelse(x >= 2 & x < 4,
                                      1/2 - x/2 + (x^2)/8,
                                      ifelse(x <= 6,
                                             3/2*x - 1/8*x^2 -7/2,
                                             1.))
    return (tr) }

library(GoFKernel) # in order to use "inverse" function
qtriangular <- Vectorize(inverse(ptriangular, lower=2, upper=6))


rtriangular <- function(n) {
    return(qtriangular(runif(n)))
}
```

```r
x <- seq(2, 6, 0.01)
y <- dtriangular(x)
options(repr.plot.width = 5, repr.plot.height = 4)
```

```r
c <- ptriangular(x)
plot(x, c, main = 'ptriangular', type = 'l', col = "blue", lwd = 2, ylab = 'ptriangular X >= x')
```

```r
r <- rtriangular(10^4)
h <- hist(r, prob=TRUE, col = "deepskyblue2", density = 30, ylim = c(0,0.5), main = 'rtriangular', xlab = '')
lines(x, y, col="red", lwd=2.5)
```

# Exercise 4

The analyzed database (https://www.gstatic.com/covid19/mobility/Region_Mobility_Report_CSVs.zip) contains files for different country related to the years 2020 and 2021.
The files contain the data related to the entire country in the first rows, then it contains the same data for all the different regions and subregions of that country in the following rows: as a consequence, in order to analyze the data of a country, we first have to merge the two files from 2020 and 2021 and then to select only the data related to the entire country, that is to say only the rows in which the columns related to regions and subregions are filled with NA values. Otherwise we would have different values for each day in the database.
Another problem is related to the mean on weeks and months: the lubridate::week and lubridate::months functions assign to each date the related week and month of the year; so, for example, the week related to the 21st of February 2020 is the same assigned to the 21st of February 2021; the same for the months. This represents a problem for computing the averages over weeks and  months, since the function "aggregate" computes them by considering the rows that have the same value in a specified columns. In order to deal with this problem, we add 12 to the months of 2021, and 53 to the weeks of 2021.

```r
library(tidyverse)
library(purrr)
library(dplyr)
library(lubridate)
```

```r
read_my_csv <- function(years, country) {
    
    filename <- paste0(years, "_", country, "_Region_Mobility_Report.csv")
    
    temp <- tempfile()
    
    download.file("https://www.gstatic.com/covid19/mobility/Region_Mobility_Report_CSVs.zip", temp)
    
    data <- read_csv(unz(temp, filename), col_types = cols())
    
    unlink(temp)
    
    return(data)
    
    }
```

```r
years <- 2020:2021

italy <- map_df(years, read_my_csv, country = "IT")
```

```r
# select only data related to the entire nation i.e. sub_region_1 = NA
italy <- italy[is.na(italy$sub_region_1),]

italyw <- italy %>%
        select(date, (retail_and_recreation_percent_change_from_baseline:residential_percent_change_from_baseline)) 

# Set Weeks number. Date already of class `Date`
italyw$weeks <- lubridate::week(italyw$date)
italyw$date <- NULL

mask.week = format(italy$date, format="%Y") == 2021
italyw$weeks[mask.week] = italyw$weeks[mask.week] + 53

# Aggregate over week number
italyw <- aggregate(.~weeks, FUN=mean, data=italyw, na.rm=TRUE)
```

```r
par(mfrow=c(3,2), oma=c(10,0,10,0))
options(repr.plot.width=20, repr.plot.height=20)

plot(italyw$weeks, italyw$retail_and_recreation_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "retail_and_recreation in Italy per week", xlab="week", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italyw$weeks, italyw$grocery_and_pharmacy_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "grocery_and_pharmacy in Italy per week", xlab="week", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italyw$weeks, italyw$parks_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "parks in Italy per week", xlab="week", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italyw$weeks, italyw$transit_stations_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "transit_stations in Italy per week", xlab="week", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italyw$weeks, italyw$workplaces_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "workplaces in Italy per week", xlab="week", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italyw$weeks, italyw$residential_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "residential in Italy per week", xlab="week", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
title("ITALY per Week", outer = TRUE, cex.main=8)
```

```r
italym <- italy %>%
        select(date, (retail_and_recreation_percent_change_from_baseline:residential_percent_change_from_baseline)) 

# Set Weeks number. Date already of class `Date`
italym$months <- lubridate::month(italym$date)
italym$date <- NULL

mask.month = format(italy$date, format="%Y") == 2021
italym$months[mask.month] = italym$months[mask.month] + 12

# Aggregate over week number
italym <- aggregate(.~months, FUN=mean, data=italym, na.rm=TRUE)
```

```r
par(mfrow=c(3,2), oma=c(10,0,10,0))

options(repr.plot.width=20, repr.plot.height=20)

plot(italym$months, italym$retail_and_recreation_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "retail_and_recreation in Italy per month", xlab="month", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italym$months, italym$grocery_and_pharmacy_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "grocery_and_pharmacy in Italy per month", xlab="month", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italym$months, italym$parks_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "parks in Italy per month", xlab="month", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italym$months, italym$transit_stations_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "transit_stations in Italy per month", xlab="month", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italym$months, italym$workplaces_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "workplaces in Italy per month", xlab="month", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
plot(italym$months, italym$residential_percent_change_from_baseline, cex.axis = 2, cex.lab=1.8, type = 'b', main = "residential in Italy per month", xlab="month", ylab = "percent change from baseline", cex.main = 2, col="darkblue", lwd = 3)
grid()
title("ITALY per Month", outer = TRUE, cex.main=8)
```
